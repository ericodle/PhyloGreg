```python
!pip install biopython

import seaborn as sns
import matplotlib.pyplot as plt
from Bio import SeqIO

# read in fasta file
fasta_file = "/content/drive/MyDrive/euk_extract.fasta"
records = SeqIO.parse(fasta_file, "fasta")

# get sequence lengths and GC content
seq_lengths = []
gc_content = []
for record in records:
    seq_lengths.append(len(record.seq))
    gc_content.append((record.seq.count("G") + record.seq.count("C")) / len(record.seq))

# create histograms
sns.set_style("darkgrid")
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))
sns.histplot(seq_lengths, ax=ax1)
ax1.set_xlabel("Sequence Length")
ax1.set_ylabel("Count")
ax1.set_title("Distribution of Sequence Lengths")

sns.histplot(gc_content, ax=ax2)
ax2.set_xlabel("GC Content")
ax2.set_ylabel("Count")
ax2.set_title("Distribution of GC Content")

plt.show()
```
